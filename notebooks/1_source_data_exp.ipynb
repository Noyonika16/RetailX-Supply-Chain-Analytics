```python
import pandas as pd

customers = pd.read_csv("../data/raw/olist/olist_customers_dataset.csv")
orders = pd.read_csv("../data/raw/olist/olist_orders_dataset.csv")
order_items = pd.read_csv("../data/raw/olist/olist_order_items_dataset.csv")
products = pd.read_csv("../data/raw/olist/olist_products_dataset.csv")
payments = pd.read_csv("../data/raw/olist/olist_order_payments_dataset.csv")
reviews = pd.read_csv("../data/raw/olist/olist_order_reviews_dataset.csv")
sellers = pd.read_csv("../data/raw/olist/olist_sellers_dataset.csv")
geolocation = pd.read_csv("../data/raw/olist/olist_geolocation_dataset.csv")
category_translation = pd.read_csv("../data/raw/olist/product_category_name_translation.csv")
```

```python
def explore(df, name):
    print("=" * 80)
    print(f"{name.upper()}")
    print("=" * 80)

    print("\nShape")
    print(df.shape)

    print("\nColumns")
    print(df.columns.tolist())

    print("\nData Types")
    print(df.dtypes)

    print("\nMissing Values")
    print(df.isnull().sum())

    print("\nDuplicate Rows")
    print(df.duplicated().sum())

    print("\nFirst Five Rows")
    display(df.head())

    print("\nSummary Statistics")
    display(df.describe(include="all"))
```

```python
explore(customers, "Customers")

explore(orders, "Orders")

explore(order_items, "Order Items")

explore(products, "Products")

explore(payments, "Payments")

explore(reviews, "Reviews")

explore(sellers, "Sellers")

explore(geolocation, "Geolocation")

explore(category_translation, "Category Translation")
```
